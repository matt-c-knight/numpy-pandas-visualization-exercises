```python
import matplotlib.pyplot as plt
```

```python
print(plt)
```

```python
my_list = [1,2,3,4,5,6,7]
y = [x**2 - x + 2 for x in my_list]


plt.plot(x,y)
plt.ylim(-1, 26)
plt.xlim(-1, 6)
plt.annotate('Origin', xy=(0,0))
plt.show()
```

```python
x = range(1,50)
def equation(n):
    return n**2-n+2
y = [equation(n) for n in x]
plt.plot(x,y)
plt.ylim(-1, 26)
plt.xlim(-1, 6)
plt.annotate('Origin', xy=(0,0))
plt.show()
```

```python
import math 
x = range(1,100)
y = [math.sqrt(n) for n in x]

plt.plot(x,y)
plt.ylim(-1, 10)
plt.xlim(-1, 100)
plt.annotate('Origin', xy=(0,0))
plt.show()
```

```python
x = range(1,10)
y = [math.pow(n,3) for n in x]
plt.plot(x,y)
plt.ylim(0, 100)
plt.xlim(0, 5)
plt.annotate('Origin', xy=(0.5,0))
plt.show()
```

```python
x = range(1,50)
y = [2*n for n in x]
plt.plot(x,y)
plt.ylim(-1, 100)
plt.xlim(-1, 60)
plt.annotate('Origin', xy=(0.5,0))
plt.show()
```
