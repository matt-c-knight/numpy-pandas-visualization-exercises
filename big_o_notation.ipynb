```python
import matplotlib.pyplot as plt

```

```python
plt.figure(figsize=(12, 12))


import math
plt.title("Big O Notation")
plt.xlabel('Elements')
plt.ylabel('Operations')
x = range(1,15)
y = [0*n + 1 for n in x]
plt.plot(x,y, label="O(1)")
plt.scatter(x,y, c='blue')
plt.ylim(-1,30)
plt.xlim(-1,20)


x= range(1,20)
y= [math.log(n) for n in x]
plt.plot(x,y, label="O(log n)")
plt.scatter(x,y, c='green')

x= range(1,20)
y= [n for n in x]
plt.plot(x,y, label="O(n)")
plt.scatter(x,y, c='red')

x= range(1,20)
y= [n * math.log(n) for n in x]
plt.plot(x,y, label="O(n log n)")
plt.scatter(x,y, c='orange')

x= range(1,20)
y= [math.sqrt(n) for n in x]
plt.plot(x,y, label="O(n^2)")
plt.scatter(x,y, c='purple')

x= range(1,20)
y= [2**n for n in x]
plt.plot(x,y, label="O(2^n)")
plt.scatter(x,y, c='brown')

x= range(1,20)
y= [math.factorial(n) for n in x]
plt.plot(x,y, label="O(n!)")
plt.scatter(x,y, c='yellow')

x= range(1,20)
y= [n**n for n in x]
plt.plot(x,y, label="O(n^n)")
plt.scatter(x,y, c='gray')

plt.legend(loc='upper right')
```
